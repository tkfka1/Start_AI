#### Telco Customer Churn Dataset  컬럼

1. CustomerID: Customer ID unique for each customer : 각 고객에게 고유한 고객 ID
2. gender: Whether the customer is a male or a female : 고객이 남성인지 여성인지
3. SeniorCitizen: Whether the customer is a senior citizen or not (1, 0) : 고객이 고령자인지 아닌지 (1, 0)
4. Partner: Whether the customer has a partner or not (Yes, No) : 고객이 파트너가 있는지 없는지 (예, 아니오)
5. Dependents: Whether the customer has dependents or not (Yes, No) : 고객이 부양 가족이 있는지 없는지 (예, 아니오)
6. Tenure: Number of months the customer has stayed with the company : 고객이 회사와 함께한 개월 수
7. PhoneService: Whether the customer has a phone service or not (Yes, No) : 고객이 전화 서비스를 가지고 있는지 없는지 (예, 아니오)
8. MultipleLines: Whether the customer has multiple lines or not (Yes, No, No phone service) : 고객이 여러 회선을 가지고 있는지 없는지 (예, 아니오, 전화 서비스 없음)
9. InternetService: Customer’s internet service provider (DSL, Fiber optic, No) : 고객의 인터넷 서비스 제공업체 (DSL, 광섬유, 없음)
10. OnlineSecurity: Whether the customer has online security or not (Yes, No, No internet service) : 고객이 온라인 보안을 가지고 있는지 없는지 (예, 아니오, 인터넷 서비스 없음)
11. OnlineBackup: Whether the customer has an online backup or not (Yes, No, No internet service) : 고객이 온라인 백업을 가지고 있는지 없는지 (예, 아니오, 인터넷 서비스 없음)
12. DeviceProtection: Whether the customer has device protection or not (Yes, No, No internet service) : 고객이 장치 보호를 가지고 있는지 없는지 (예, 아니오, 인터넷 서비스 없음)
13. TechSupport: Whether the customer has tech support or not (Yes, No, No internet service) : 고객이 기술 지원을 받고 있는지 없는지 (예, 아니오, 인터넷 서비스 없음)
14. StreamingTV: Whether the customer has streaming TV or not (Yes, No, No internet service) : 고객이 스트리밍 TV를 가지고 있는지 없는지 (예, 아니오, 인터넷 서비스 없음)
15. StreamingMovies: Whether the customer has streaming movies or not (Yes, No, No internet service) : 고객이 스트리밍 영화를 가지고 있는지 없는지 (예, 아니오, 인터넷 서비스 없음)
16. Contract: The contract term of the customer (Month-to-month, One year, Two years) : 고객의 계약 기간 (월간, 1년, 2년)
17. PaperlessBilling: Whether the customer has paperless billing or not (Yes, No) : 고객이 종이 없는 청구를 받고 있는지 없는지 (예, 아니오)
18. PaymentMethod: The customer’s payment method (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic)) : 고객의 결제 방법 (전자 체크, 우편 체크, 은행 이체 (자동), 신용카드 (자동))
19. MonthlyCharges: The amount charged to the customer monthly : 고객에게 월별로 청구되는 금액
20. TotalCharges: The total amount charged to the customer : 고객에게 청구된 총 금액
21. Churn: Whether the customer churned or not (Yes or No) : 고객이 이탈했는지 아닌지 (예 또는 아니오)

#### TotalCharges 컬럼 타입 변경하기

```python
# TotalCharges : 월사용요금, 실수형으로 보인다.
df['TotalCharges']
```

```python
# TotalCharges 컬럼 타입을 float로 변경해 보자.
# 문자열을 숫자형으로 변경할수 없으므로 에러 발생

df['TotalCharges'].astype(float)
```

```python
# Boolean indexing으로 검색
(df['TotalCharges'] == '') | (df['TotalCharges'] == ' ')
```

```python
# Boolean indexing으로 검색

## df['TotalCharges'] ==  '' 빈 str or ' ' 띄어 쓰기 

cond = (df['TotalCharges'] == '') | (df['TotalCharges'] == ' ')
df[cond]
```

####  
##### <font color=blue> **[문제] df 데이터프레임의 'TotalCharges' 컬럼의 값 ' ' --> '0' 변경하세요.** </font>

```python
# DataFrame replace 함수 
# 대상 컬럼 : 'TotalCharges'  

## replace == > ' ' 는 ['0'] 으로 변환 inplace -> 변환시키면 그대로 넣기

df['TotalCharges'].replace([' '], ['0'], inplace=True)
```

####  
##### <font color=blue> **[문제] df 데이터프레임의 'TotalCharges' 컬럼 타입을 object에서 float 변경하세요.** </font>

```python
# 'TotalCharges' 컬럼 type을 float로 변경
# 결과를 TotalCharges 컬럼에 다시 넣어야 합니다.


df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
# 다시 Boolean indexing으로 검색 : '' 값을 가진 Row 없음

cond = (df['TotalCharges'] == '') | (df['TotalCharges'] == ' ')
df[cond]
```

```python
# TotalCharges 컬럼 float 변경 확인 
df.info()
```
